```python
#import libraries 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#load data

df = pd.read_csv(r'C:\Users\issas\Downloads/moviesdata.csv', encoding = 'unicode_escape')
```

```python
#view data

df.head()
```

```python
#check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#check data types 

df.dtypes
```

```python
#change data types

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df.head()
```

```python
#create new correct year column

df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
#sort values by gross

df= df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#to view all the data
pd.set_option('display.max_rows', None)

df.head()
```

```python
#check duplicates and drop duplicates

df['company'].drop_duplicates().sort_values(ascending = False)

df.head()
```

```python
#create a scatter plot with budget vs gross

plt.scatter(x= df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```python
df.head()
```

```python
#create a regression plot of budget vs gross

sns.regplot(x= 'budget', y='gross', data= df, scatter_kws= {"color": "red"}, line_kws= {"color": "blue"})
```

```python
#check correlation
#pearson, kendall, spearman
df.corr(method= 'pearson')
```

```python
#high correlation between budget and gross 
```

```python
#visualize correlation

correlation_matrix= df.corr()
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
plt.show()
```

```python
#look at company 

df_numerized =df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix= df_numerized.corr()
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
plt.show()
```

```python
correlation_mat= df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head(10)
```

```python
#sort correlation values
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head(10)
```

```python
#select and sort by highest correlated pairs

high_corr= sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
